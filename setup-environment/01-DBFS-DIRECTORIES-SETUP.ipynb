```python
# Databricks notebook
# 01-dbfs-directory-setup
# Purpose: Create the initial file structure for the SUSEP Lakehouse project using Unity Catalog Volumes

from datetime import datetime

# ============================================================
# 1. Project configuration
# ============================================================

project_name = "lakehouse-susep"

# First, check available catalogs
display(spark.sql("SHOW CATALOGS"))
```

```python
# ============================================================
# 2. Unity Catalog / Volume configuration
# ============================================================

catalog_name = "workspace"        
schema_name = "susep"
schema_name_bronze = "susep_bronze"
schema_name_silver = "susep_silver"
schema_name_gold = "susep_gold"
volume_name = "lakehouse"

# Create schema/database if it does not exist
spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog_name}`.`{schema_name_bronze}`")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog_name}`.`{schema_name_silver}`")
spark.sql(f"CREATE SCHEMA IF NOT EXISTS `{catalog_name}`.`{schema_name_gold}`")

# Create volume if it does not exist
spark.sql(f"""
CREATE VOLUME IF NOT EXISTS `{catalog_name}`.`{schema_name_bronze}`.`{volume_name}`
""")

# Volume root path
base_path = f"/Volumes/{catalog_name}/{schema_name_bronze}/{volume_name}/{project_name}"

print(f"Project root path: {base_path}")
```

```python
# ============================================================
# 3. Directory structure
# ============================================================

directories = [
    base_path,

    # Main Lakehouse layers
    f"{base_path}/bronze",


    # Bronze - raw landing area
    f"{base_path}/bronze/raw",
    f"{base_path}/bronze/raw/zip",
    f"{base_path}/bronze/raw/csv",
]

for directory in directories:
    dbutils.fs.mkdirs(directory)
    print(f"Created or already exists: {directory}")
```

```python
# ============================================================
# 4. Validate structure
# ============================================================

print("Root directory content:")
display(dbutils.fs.ls(base_path))

print("Bronze directory content:")
display(dbutils.fs.ls(f"{base_path}/bronze"))
```
