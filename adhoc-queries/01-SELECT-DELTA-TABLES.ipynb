```python
%sql
select * from susep_bronze.aux_de_para_grupo_empresa
where coenti in (select coenti from susep_bronze.aux_de_para_grupo_empresa group by coenti having count(*) > 1)
order by 1
```

```python
%sql
select distinct coramo, cod_ramo_susep, nome_ramo_susep  
from susep_bronze.ses_seguros seg
 JOIN susep_bronze.aux_de_para_ramo_susep ramo
 on seg.coramo = substring(ramo.cod_ramo_susep,3,2) 
where cast(coramo as bigint) not in (select cast(cod_ramo_susep as bigint) from susep_bronze.aux_de_para_ramo_susep)
order by 1 
```

```python
%sql
select distinct coramo
from susep_bronze.ses_seguros seg
 where coramo not in (select cod_ramo_susep from susep_bronze.aux_de_para_ramo_susep)

order by 1 desc
```

```python
%sql
select distinct coramo
from susep_bronze.ses_seguros seg
 where coramo not in (select cod_ramo_susep from susep_bronze.aux_de_para_ramo_susep)
```
